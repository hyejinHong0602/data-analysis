## Regression Analysis

### Drop Columns

```python
airbnb.drop(['host_id','latitude','longitude','neighbourhood','number_of_reviews','reviews_per_month'], axis=1, inplace=True)
#examing the changes
airbnb.head(5)
```

```python
#Encode the input Variables
def Encode(airbnb):
    for column in airbnb.columns[airbnb.columns.isin(['neighbourhood_group', 'room_type'])]:
        airbnb[column] = airbnb[column].factorize()[0]
    return airbnb

airbnb_en = Encode(airbnb.copy())

## 범주형 변수가 문자열 data type을 가지고 있을 경우, 클래스 레이블을 정수형으로 인코딩 해야한다.
## factorize()는 2가지 값을 반환함. [0] : 정수로 인코딩 된 값. [1] : 인코딩된 범주.

## factorize() 예시. 
#factorize == 
# 0번째는 index ex) b,b,a,c,b == 0,0,1,2,0
# 1번째는 b,a,c 
```

```python
airbnb_en.head(15)
```

```python
#Get Correlation between different variables
corr = airbnb_en.corr(method='kendall')  ## 상관계수 가져오기
plt.figure(figsize=(18,12))
sns.heatmap(corr, annot=True)  ## 히트맵으로 보여주기 & 값 써주기 (annot)
airbnb_en.columns
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import r2_score
```

```python
#Defining the independent variables and dependent variables ## 독립변수와 종속변수 정의
x = airbnb_en.iloc[:,[0,1,3,4,5]]  ## 0,1,3,4,5번째 열만 넣기. (price 빼고)
y = airbnb_en['price']

#Getting Test and Training Set
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.1,random_state=353) 
## train_test_split : train set과 validation set 분리
## test_size : 테스트셋 구성비율 : 0.1을 test(validation)셋으로 구성 (근데 너무 낮아서 overfitting 되어보임..)
## random_state=353 : 고정값을 줘서 매번 데이터셋이 변경되는 걸 방지한다. (여기 숫자는 딱히 의미있는 건 아님. random seed에 있는 숫자와 비슷한 느낌.)

x_train.head()
y_train.head()
```

```python
x_train.shape
```

```python
#Prepare a Linear Regression Model
reg=LinearRegression() ## 선형회귀모델 생성.
reg.fit(x_train,y_train)  ##fit 함수 통해서 선형회귀모델에 필요한 두 변수 전달
y_pred=reg.predict(x_test)  ## 새로운 x값을 넣어 y값을 예측할 수 있게 됨.
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)  ## 결정계수 출력
```

```python
#Prepairng a Decision Tree Regression 
from sklearn.tree import DecisionTreeRegressor
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.1,random_state=105)
DTree=DecisionTreeRegressor(min_samples_leaf=.0001)
## min_samples_leaf = .0001 (default는 1) : 잎노드가 되려면 가지고 있어야 할 최소 샘플 수
## 클수록 가지치기 (과대적합 방지) / 작을수록 정확하게 (과대적합)

DTree.fit(x_train,y_train)
y_predict=DTree.predict(x_test)
from sklearn.metrics import r2_score
r2_score(y_test,y_predict)
```

```python
#Prepairng a Decision Tree Regression
from sklearn.tree import DecisionTreeRegressor
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.1,random_state=105)
DTree=DecisionTreeRegressor(min_samples_leaf=.0001)
DTree.fit(x_train,y_train)
y_predict=DTree.predict(x_test)
from sklearn.metrics import r2_score
r2_score(y_test,y_predict)
```
